```python
import pandas as pd
import numpy as np
from nba_api.stats.static import players,teams
from nba_api.stats.endpoints import playercareerstats,playergamelog
from nba_api.stats.endpoints import ShotChartDetail
import requests
import os.path

def find(data,key,value):
    for i in data:
        
        
        if int(i[key]) == int(value):
                
            return i

    return('Not found')
def get_shotdata(player_id,seasons,stype ='Regular Season',term = '',league_id = None):
    shots = pd.DataFrame()
    ranges = ['16-24 ft.', '16-24 ft.', '16-24 ft.', '16-24 ft.', '16-24 ft.', '24+ ft.', '24+ ft.', '24+ ft.', '24+ ft.', '24+ ft.', '8-16 ft.', '8-16 ft.', '8-16 ft.', 'Back Court Shot', 'Less Than 8 ft.']
    
    areas = ['Center(C)', 'Left Side Center(LC)', 'Left Side(L)', 'Right Side Center(RC)', 'Right Side(R)', 'Center(C)', 'Left Side Center(LC)', 'Left Side(L)', 'Right Side Center(RC)', 'Right Side(R)', 'Center(C)', 'Left Side(L)', 'Right Side(R)', 'Back Court(BC)', 'Center(C)']
    carry =''
    #print(season)
    context = 'FGA'
    map_type =''
    if term == 'vs':
        context='OPP_FGA'
        
    elif term =='on':
        context = 'TM_FGA'
    else:
        print('failed to save')
        break
      
    if league_id == None:
        league_id = '00'
    if stype =='Playoffs':
        season_index = 2
        carry = 'ps'
        #print('playoffs')
    else:
        season_index = 0
        
    season = seasons[0]
    season = int(season[0:4])+1
    path = str(season)
    if stype =='Playoffs':
        path+='ps'
    file = player_id+term+'.csv'
    if os.path.exists(path):
        if file in os.listdir(path):
            return
    #print( find(nba_players,'full_name',player_name))
    #team_id = commonplayerinfo.CommonPlayerInfo(player_id).get_data_frames()[0]['TEAM_ID'].values
    #team_id = find(nba_teams,'abbreviation',team)['id']
    #print(player_id)
    #print(season)
    #logs = playergamelogs.PlayerGameLogs(player_id_nullable=player_id,season_nullable=season).get_data_frames()[0]
    career = playercareerstats.PlayerCareerStats(player_id=player_id)
    #print(career.get_data_frames())
    career = career.get_data_frames()[season_index]
    #print(career)

    career = career[career.SEASON_ID.isin(seasons)]
    #print(career)
    #print(len(career))
    if len(career.index) == 0:
        return shots
    seasonfga = career['FGA'].sum()
    team_id = list(career['TEAM_ID'].unique())
    #print(team_id)
    #print(seasons)
    #print(team_id)
    #print(team_id)
    #team_id = [1610612744]
    
    playershots = []
    #print(seasons)
    for season in seasons:
        
        for i in range(len(team_id)):
            #print(player_id)
            #print(team_id[i])
            #print(season)
            #print(stype)
            #print(league_id)
            shot_chart = ShotChartDetail(player_id=player_id,
                          team_id=team_id[i],
                          season_nullable= season,
                          season_type_all_star=stype, 
                          context_measure_simple = context,
                                        league_id = league_id).get_data_frames()
            #print('generates shotchart')
            #print(shot_chart)
            #print(shot_chart[1]['FGA'].sum())
            if i ==0:
                df_player_shots = shot_chart[0]

            else:
                df_player_shots= pd.concat([df_player_shots,shot_chart[0]])
        playershots.append(df_player_shots)
        #print('added playershots')

        #print(df_league_average)
    
    shotchart= pd.concat(playershots)
    season = seasons[0]
    season = int(season[0:4])+1
    path = str(season)
    if stype =='Playoffs':
        path+='ps'
    # Check whether the specified path exists or not
    isExist = os.path.exists(path)
    if not isExist:

   # Create a new directory because it does not exist
        os.makedirs(path)
        print('Making Folder ' +path)
    shotchart.to_csv(path+'/'+str(player_id)+term+'.csv',index = False)
    #print(len(shotchart))

    #print(df_league_average)
    #print(shotchart)
    #print(df_league_average)
    #print(df_player_shots['SHOT_ATTEMPTED_FLAG'].sum())
'''
    shotdiagram = shotchart.groupby(['SHOT_ZONE_RANGE','SHOT_ZONE_AREA','SHOT_MADE_FLAG']).size().unstack(fill_value=0).reset_index()
    #print('made it to shot diagram')
    if 0 not in shotdiagram.columns:
        #print('No Misses')
        #print(shotchart)
        #print(shotdiagram)
        shotdiagram[0] = 0
    if 1 not in shotdiagram.columns:
        #print('No Makes')
        #print(shotchart)
        #print(shotdiagram)
        shotdiagram[1] = 0
    if len(shotdiagram)==1:
        print('1 shot')
        #print(player_id)
        #print(shotdiagram)
        
        
    
    
    
    
    
    shots[['SHOT_ZONE_RANGE', 'SHOT_ZONE_AREA', 'Missed','Made']] = shotdiagram[['SHOT_ZONE_RANGE', 'SHOT_ZONE_AREA', 0, 1]]
    
    #print(shots)

    frameshots=shots['SHOT_ZONE_RANGE'].tolist()
    frameareas=shots['SHOT_ZONE_AREA'].tolist()
    
    
    zippedshots = list(zip(frameshots,frameareas))
    
    shots = shots.reset_index(drop = True)
    #print('reset shot map')
    #print(shots)
    
    for item in list(zip(ranges,areas)):
        
        if item not in zippedshots:
            #blank_shots = pd.DataFrame({'SHOT_ZONE_RANGE': item[0], 'SHOT_ZONE_AREA':item[1],'Made': 0, 'Missed': 0})
            #shots = pd.concat([shots,blank_shots])
            shots.loc[len(shots.index)] = [item[0],item[1],0,0] 

    
    #print('added empty rows to shotmap')
    #print(shots)
    shots = shots.sort_values(by=['SHOT_ZONE_RANGE','SHOT_ZONE_AREA'])
    shots = shots.reset_index(drop = True)
    shotmapfga = shots['Made'].sum() + shots['Missed'].sum()
    #print(shotmapfga)
    if (seasonfga!=shotmapfga):
        print('not aligned')

        #print(career)
        #print(shotchart)
        print('Career based request returns ' +str(seasonfga) +' attempts') 
        print('Shotmap based request returns ' +str(shotmapfga) +' attempts') 
        print('There are '+str(len(shotchart))+ ' fga in the shotchart')

        #print(shots)
        #print(shotdiagram)
        
    
    #print(shots.columns)
    #print(shotdiagram)

    #shots = shots.reset_index(drop =True)
    #print(shotdiagram)
    #print(shots)

    #print(shots)
    return shots
'''
df = pd.read_csv('player_index.csv')
#df = df[df.year<2010]
#df = df[df.year!=2010]
df['Season'] = (df['year']-1).astype(str)+'-'+df['year'].astype(str).str[2:4]


seasons = df['Season'].unique()


#seasons = seasons[7:]
print(seasons)



#df= df[df.year==2010]
def year_data(df,season,averages,count=0,ps = False):
    if ps == False:
        stype='Regular Season'
    else:
        stype='Playoffs'
    all_averages=[]

    df = df.iloc[count:]
    nba_id = df['PLAYER_ID'].tolist()

    players= df['PLAYER_NAME'].tolist()
    count = count
    not_found = []
    for nba_id,player in zip(nba_id,players):
        #print(player)
        #print(season)
        count+=1

        #print(role)
        #print(season)
        nba_id = str(int(nba_id))
        try:
            get_shotdata(nba_id,[season],term='vs',stype=stype)
            get_shotdata(nba_id,[season],term='on',stype=stype)
            #print(shots)

        except ValueError: 
            not_found.append([player,nba_id,season])
            print(player +' with id '+ nba_id+ ' in season '+season +' not found')
        except requests.Timeout as err:
            not_found.append([player,nba_id,season])
            print('The count when timing out was ' +str(count))
            print(player +' with id '+ nba_id+ ' in season '+season +' not found')


              
        if count %10 == 0:
            print(str(count) + ' rows processed')
    return averages,not_found
#seasons = ['2009-10']
missing = {}
count = 0
def set_season(df,seasons,count,ps = False):
    all_averages= []
    trail = ''
    if ps == True:
        trail='ps'
        

    for season in seasons:
        averages = {}
        year = int(season[0:4]) +1
        year_df = df[df.year == year]



        if os.path.isfile('missing_'+str(year)+trail+'.csv'):
            missing[year] = pd.read_csv('missing_'+str(year)+trail+'.csv')

        averages,not_found = year_data(year_df,season,averages,count = count,ps = ps)
        missing[year] = pd.DataFrame(not_found,columns =['Player','NBA_ID','SEASON'])

        missing[year].to_csv('missing_'+str(year)+trail+'.csv',index = False)
        count = 0
        print(missing[year])
        print('Season Completed '+str(year))
        all_averages.append(averages)
    return all_averages
#averages = set_season(df,seasons,count,ps=False)   
#print(averages)
```
