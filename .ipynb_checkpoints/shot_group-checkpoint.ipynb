```python

def on_maps(path,term):
    onmaps = [file for file in os.listdir(path) if term in file and '.csv' in file]
    len(onmaps)
    frames = []
    for file in onmaps:
        df = pd.read_csv(path+'/'+file)
        player_id = file[:-6]
        if (len(player_id)) == 0:
            print(player_id)
        df['SHOT_ID'] = df['GAME_ID'].astype(str)+df['GAME_EVENT_ID'].astype(str)
        df['MAP_ID'] = player_id
        frames.append(df)

    total = pd.concat(frames)
    total = total.dropna(subset=['MAP_ID'])
    on_dict = {}

    for index, row in total.iterrows():
        shot_id = row['SHOT_ID']
        if shot_id not in on_dict:
            on_dict[shot_id] = row['MAP_ID']
        else:
            if row['MAP_ID'] not in on_dict[shot_id]:
                on_dict[shot_id] +=('|'+row['MAP_ID'])
    cleaned = total.drop(columns = ['MAP_ID'])

    cleaned = cleaned.drop_duplicates(subset='SHOT_ID')

    cleaned['PLAYERS_ON'] = cleaned['SHOT_ID'].map(on_dict)
    return cleaned
for i in range(2024,2025):
    path = str(i)
    term = 'on'
    cleaned = on_maps(path,term)
    cleaned.to_csv('full/'+path+'.csv',index = False)
    print(path+ term+' done')
    #path = str(i)+'ps'
    term = 'vs'
    cleaned = on_maps(path,term)
    cleaned.to_csv('full/'+path+'vs.csv',index = False)
    print(len(cleaned))
    print(path+term+ ' done')
```
